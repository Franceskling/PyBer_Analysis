```python
# Import dependencies.
import matplotlib.pyplot as plt

```

```python
#Data for line graphs 

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot

plt.plot(x_axis,y_axis)
```

```python
#Create the plot with ax.plt()

fig,ax=plt.subplots()
ax.plot(x_axis,y_axis)

```

```python
#Another way to create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
#Use object oriented approach to do the following...
#Change marker to a diamond, Change the line color to green, change line width to 2 points
fig,ax=plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2)
#Add a legend for the city of Boston.
ax.legend('Boston')
#Add a title and axes labels.
ax.set_title('PyBer Fare per Month')
#Add a y-axis limit.
ax.set_ylim(0,50)
#Add grid lines.
ax.grid()
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
#plot line chart with error bars
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
#add a cap to the error bars
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#using object oriented
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#Error bars on a bar chart
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#object oriented 
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#Adding ticks
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
import numpy as np
np.arange(0, 51, step=5.0)
```

```python
#adding ticks object oriented
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
#adding minor ticks object oriented
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
